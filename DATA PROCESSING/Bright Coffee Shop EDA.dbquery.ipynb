```sql
---------------TABLE OVERVIEW
SELECT *
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
LIMIT 10;
--OUTPUT: Returns all colums with 10 rows
---------------DATA OVERVIEW
SELECT COUNT(*) AS Total_Number_of_Rows, 
        COUNT(DISTINCT transaction_id) AS User_ID
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: Starts counting Total Number of Rows and then Counts Distinct number of Rows
--Number of Rows/Duplicates = 149116 (single transaction per customer)

---------------DATA DURATION
SELECT
    MIN(transaction_date) AS Start_date,
    Max(transaction_date) AS End_date
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: Shows the duration of the data which is 6 months worth of data
--2023-01-01 to 2023-06-30

--------------CHECKING FOR NULLS, ''
SELECT 
COUNT(CASE WHEN transaction_id IS NULL THEN 1 END) AS User_ID,
COUNT(CASE WHEN transaction_date IS NULL THEN 1 END) AS Date,
COUNT(CASE WHEN transaction_time IS NULL THEN 1 END) AS Transactyion_Time,
COUNT(CASE WHEN transaction_qty IS NULL THEN 1 END) AS Quantity,
COUNT(CASE WHEN store_id IS NULL THEN 1 END) AS Store_ID,
COUNT(CASE WHEN store_location IS NULL THEN 1 END) AS Location,
COUNT(CASE WHEN product_id IS NULL THEN 1 END) Product_ID,
COUNT(CASE WHEN unit_price IS NULL THEN 1 END) AS Price_Unit,
COUNT(CASE WHEN product_category IS NULL THEN 1 END) AS Product_Category,
COUNT(CASE WHEN product_type IS NULL THEN 1 END) AS Product_Type,
COUNT(CASE WHEN product_detail IS NULL THEN 1 END) AS Product_Detail
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: We have zero null values

-----------------------CHECKING FOR EMPTY VALUES ' '
SELECT
SUM(CASE WHEN store_location = ' ' THEN 1 ELSE 0 END) AS Location,
SUM(CASE WHEN product_category = ' ' THEN 1 ELSE 0 END) AS Product_Category,
SUM(CASE WHEN product_type = ' ' THEN 1 ELSE 0 END) AS Product_Type,
SUM(CASE WHEN product_detail = ' ' THEN 1 ELSE 0 END) AS Product_Detail
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: We do not have columns containing Empty ' ' values
--NOTE: Only string columns can contain ' ' values. Numeric/date/timestamp columns can only be NULL or valid values.

---------------------------DATA CLEANING------------------------
----------------KEY PERFORMANCE INDICATORS-----------
---------------------------TOTAL REVENUE
SELECT
    SUM(transaction_qty * unit_price) AS Total_Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: Total Revenue for 6 month is R698,812.33

----------------------------NUMBER OF Transactions
SELECT
   COUNT(DISTINCT transaction_id) AS Number_of_Transactions
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: We have 149116 customers

----------------------------NNUMBER OF STORES
SELECT 
    COUNT(DISTINCT store_location) AS Stores
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: We have 3 stores

----------------------------NUMBER OF PRODUCTS
SELECT 
    COUNT(DISTINCT product_type) AS Stores
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--Returns Only the Total
SELECT
    product_type AS Products,
    COUNT(DISTINCT product_type) AS Number_of_Products
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_type;
--Returns Distinct product and its Count
--OUTPUT: We have 29 types of products

------------------------------PRODUCT SEGMENTS
SELECT
    product_category AS Product_Segment,
    COUNT(DISTINCT product_type) AS Number_of_Products
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_category;
-- OUTPUT: We have 9 Product Segments
---------------------------AVERAGE TRANSACTIONS
SELECT
    ROUND(
    AVG(unit_price * transaction_qty),2
    ) AS Average_Transaction_Value
FROM bright_coffee_shop.default.bright_coffee_shop_analysis;
--OUTPUT: On an average 5 (4.69) transactions are made on a product

------------------------ANALYSIS---------------------------
------------------------REVENUE TREND OVERTIME
SELECT
    transaction_date,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY transaction_date
ORDER BY transaction_date;
--OUTPUT: Shows how revenue changes daily over the 6-month period.

SELECT
    MONTH(transaction_date) AS Month_Number,
    DATE_FORMAT(transaction_date, 'MMMM') AS Month,
    ROUND(SUM(transaction_qty * unit_price), 2) AS Total_Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY
    MONTH(transaction_date),
    DATE_FORMAT(transaction_date, 'MMMM')
ORDER BY Month_Number;
--OUTPUT: It shows each month and its totals


------------------------REVENUE BY PRODUCT SEGMENT
SELECT
    product_category AS Product_Segment,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_category
ORDER BY Revenue DESC;
--OUTPUT: Shows Which product categories generate the most money

-----------------------TOP 10 PRODUCTS BY REVENUE
SELECT
    product_type,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_type
ORDER BY Revenue DESC
LIMIT 10;
--OUTPUT: Shows the products that contribute most to company revenue.

SELECT
    transaction_date,
    product_type,
    ROUND(SUM(transaction_qty * unit_price), 2) AS Daily_Product_Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY
    transaction_date,
    product_type
ORDER BY
    transaction_date,
    Daily_Product_Revenue DESC;
--OUTCOME: Shows revenue generated by each product daily
-----------------------TOP 10 BOTTOM PRODUCTS BY REVENUE
SELECT
    product_type,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_type
ORDER BY Revenue ASC
LIMIT 10;
--OUTPUT: Shows underperforming products

------------------------TIME BUCKET

SELECT
    CASE
        WHEN HOUR(transaction_time) BETWEEN 6 AND 11 THEN 'Morning'
        WHEN HOUR(transaction_time) BETWEEN 12 AND 16 THEN 'Afternoon'
        WHEN HOUR(transaction_time) BETWEEN 17 AND 21 THEN 'Evening'
    ELSE 'Night'
END AS Time_Bucket,

ROUND(
    SUM(transaction_qty * unit_price),
    2
) AS Total_Revenue

FROM bright_coffee_shop.default.bright_coffee_shop_analysis

GROUP BY

    CASE
        WHEN HOUR(transaction_time) BETWEEN 6 AND 11 THEN 'Morning'
        WHEN HOUR(transaction_time) BETWEEN 12 AND 16 THEN 'Afternoon'
        WHEN HOUR(transaction_time) BETWEEN 17 AND 21 THEN 'Evening'
    ELSE 'Night'
END
ORDER BY Total_Revenue DESC;

--THEN
--SELECT
  --  Time_Bucket,
    --SUM(transaction_qty * unit_price) AS Revenue
--FROM bright_coffee_shop.default.bright_coffee_shop_analysis
--GROUP BY Time_Bucket
--ORDER BY Revenue DESC;

--------------------------HOUR PEAK SALES 
SELECT
    HOUR(transaction_time) AS Sales_Hour,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY Sales_Hour
ORDER BY Revenue DESC;
--OUTPUT: Shows hour of day where we make most money
--Exact high-performing hours.

-------------------------STORE PERFORMANCE
SELECT
    store_location,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY store_location
ORDER BY Revenue DESC;
--OUTPUT: Shows which store contributes most revenue.

----------------------PRODUCT QUALITY vs REVENUE
SELECT
    product_type,
    SUM(transaction_qty) AS Units_Sold,
    SUM(transaction_qty * unit_price) AS Revenue
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_type;
--OUTPUT: Shows products that sell a lot versus products that generate high revenue.

SELECT
    product_type,
    ROUND(AVG(unit_price * transaction_qty),2)
    AS Average_Transaction_Value
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY product_type
ORDER BY Average_Transaction_Value DESC;
--OUTPUT: Shows which products increase customer spending.

--------------------------AVERAGE DAILY TRANSACTION
SELECT
    transaction_date,
    ROUND(SUM(unit_price * transaction_qty) 
    / COUNT(transaction_id),2 
) AS Average_Daily_Transaction_Value
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
GROUP BY transaction_date
ORDER BY transaction_date;
--OUTPUT: It shows daily transactions on average

--cte

WITH  cte AS (

SELECT

    transaction_id AS TRANSACTION_ID,
    transaction_date AS DATE,
    transaction_time AS TRANSACTION_TIME,
    store_location AS STORE,
    product_category AS PRODUCT_CATEGORY,
    product_type AS PRODUCT_TYPE,
    product_detail AS PRODUCT_DETAIL,
    unit_price PRICE_PER_UNIT,
    transaction_qty AS QUANTITY,

-- Revenue

(unit_price * transaction_qty) AS TOTAL_AMOUNT,


-- Date Analysis

    YEAR(transaction_date) AS SALES_YEAR,
    MONTH(transaction_date) AS SALES_MONTH,
    DATE_FORMAT(transaction_date,'MMMM') AS MONTH_NAME,
    DAYNAME(transaction_date) AS DAY_NAME,
--NOTE:Sales MONTH is not really necessary in this regard, it is to show the ability to use date functions

    CASE
        WHEN DAYOFWEEK(transaction_date) IN (1,7)
        THEN 'Weekend'
        ELSE 'Weekday'
        END AS DAY_TYPE,


-- Time Analysis

    HOUR(transaction_time) AS SALES_HOUR,

    CASE

        WHEN HOUR(transaction_time) BETWEEN 6 AND 11 THEN 'Morning'
        WHEN HOUR(transaction_time) BETWEEN 12 AND 16 THEN 'Afternoon'
        WHEN HOUR(transaction_time) BETWEEN 17 AND 21 THEN 'Evening'
        ELSE 'Night'
        END AS TIME_BUCKET
FROM bright_coffee_shop.default.bright_coffee_shop_analysis
)
SELECT *
FROM cte;


```
